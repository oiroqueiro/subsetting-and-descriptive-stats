#### Select all the customers from Spain.
**Hint**: Remember that you are not asked to find orders from Spain but customers. A customer might have more than one order associated. 

```python
# your code here

orders[orders['Country'] == 'Spain']['CustomerID'].unique()
```

#### How many customers do we have in Spain?

```python
# your code here    

len(set(orders[(orders['Country'] == 'Spain')]['CustomerID']))
```

#### Select all the customers who have bought more than 50 items.
**Hint**: Remember that you are not asked to find orders with more than 50 items but customers who bought more than 50 items. A customer with two orders of 30 items each should appear in the selection.

```python
# your code here

orders.groupby('CustomerID').filter(lambda x: x['Quantity'].sum()>50)['CustomerID'].unique()
```

```python
i = 0
for customer in orders.groupby('CustomerID').filter(lambda x: x['Quantity'].sum()>50)['CustomerID'].unique():
    print(customer)
    i+=1
print(f'\nTotal: {i}')
```

#### Select orders from Spain that include more than 50 items.

```python
# your code here

orders[(orders['Country'] == 'Spain')].groupby('InvoiceNo').filter(lambda x: x['Quantity'].sum()>50)['InvoiceNo'].unique()
```

#### Select all free orders.

```python
# your code here

orders.groupby('InvoiceNo').filter(lambda x: x['UnitPrice'].sum() == 0)['InvoiceNo'].unique()
```

#### Select all orders whose description starts with `lunch bag`.

#### Select all `lunch bag` orders made in 2011.

#### Find the number of orders made by each country in the month of August.
**Hint**: Use value_counts().

```python
# your code here

"""
I'm not sure about this, I wanted to get something like this: 
2 orders from Australia:
len(orders[(orders['Country']=='Australia') & (orders['month']==8)]['InvoiceNo'].unique())

but when do the group by, and using the value_counts() always include the InvoiceNo which is what I want to count
"""

#This is the final approach
display(orders[orders['month']==8].groupby('Country')['InvoiceNo'].value_counts())
#These are other tests
display('')
display(orders[orders['month']==8].groupby('Country').value_counts(subset=['InvoiceNo']))
display('')
display(orders[orders['month']==8].groupby('Country').value_counts())
```
